# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
ad=input()
fiyat=input()
with open ("urunler.txt","w") as f:
    f.write(f"{ad}\n{fiyat}\n")
with open ("urunler.txt","r") as f:
    a=f.read()
    print(a)   
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(self,a,b,c):
        self.name=a
        self.speed=b
        self.mileage=c
bus=Vehicle("School Volvo","180","12")
print(f"Name: {bus.name} Speed: {bus.speed} Mileage: {bus.mileage}")
````
