# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

kanallar={"a":"1","b":"2","c":"3","d":"4","e":"5"}

class Kumanda:


    def __init__(self,durum="kapalı",sesi=0,kanal_listesi=["a","b","c","d","e"],kanal_sayisi=5,mevcut_kanal=1):
        self.durum=durum
        self.sesi=sesi
        self.kanal_listesi=kanal_listesi
        self.kanallar=kanallar
        self.kanal_sayisi=kanal_sayisi
        self.mevcut_kanal=mevcut_kanal


    def tv_ac(self):
        self.durum="kapalı"
        print("kumanda açılıyor...")
        t=time.ctime()
        print("Saat,",t)
        self.durum="açık"


    def tv_kapa(self):
        if self.durum=="açık":
            print("kumanda kapatılıyor...")
            self.durum="kapalı"
        else:
            print("Kumanda zaten kapalı")


    def ses(self):
        ses_son=int(input("sesi ne kadar artırmak/azaltmak istiyorsunuz?"))
        if ses_son<0:
            self.sesi+=ses_son
        else:
            self.sesi+=ses_son
        print({self.sesi})


    def kanal_ekle(self):
        if self.durum=="kapalı":
            print("Önce kumandayı açınız...")

        else:
            kanal_isminiz=input("kanal ismi giriniz")
            self.kanal_sayisi+=1
            print(f"{kanal_isminiz} isimli kanal, {self.kanal_sayisi} nolu kanala kaydedilmiştir.")

    def kanal_sayısı(self):
        print(f"{self.kanal_sayisi}")


    def rand_kanal(self):
        a=random.randint(1,self.kanal_sayisi)
        self.mevcut_kanal+=a
        print(f"{self.mevcut_kanal} nolu kanal açılıyor...")


    def tv_info(self):
        print(f"Kanallar ve numaraları: {kanallar}")

print("""
Kumanda
1. Tv aç
2. Tv kapat
3. Ses aç
4. Kanal ekle
5. Kanal sayısı öğren
6. Random kanal ekle
7. Televizyon bilgileri
Çıkmak için 'q' ya basın.""")

kumanda=Kumanda()

while True:
    islem = input("işlemi seçiniz:")
    if (islem == "q"):
        print("Kumanda kapandı")
        break
    elif (islem == "1"):
        kumanda.tv_ac()
    elif (islem == "2"):
        kumanda.tv_kapa()
    elif (islem == "3"):
        kumanda.ses()
    elif (islem == "4"):
        kumanda.kanal_ekle()
    elif (islem == "5"):
        kumanda.kanal_sayısı()
    elif (islem == "6"):
        kumanda.tv_info()
    else:
        pass
```
