# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__ (self,brand,year):
        self.brand=brand
        self.year=year

class Bus(Vehicle):
    def __init__(self,brand,year,capacity):
        super().__init__(brand,year)
        self.capacity=capacity
    
    def intro (self):
        print(f"Araba markası: {self.brand} Yılı: {self.year} Yer Sayısı: {self.capacity}")

b1=Bus("Karsan","2022","50")
Bus.intro(b1)
````
